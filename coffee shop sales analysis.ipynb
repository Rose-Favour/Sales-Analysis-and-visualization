```python
import pandas as pd

```

```python
df = pd.read_csv('coffeeShop.csv')
```

```python
df.head()
```

```python
dfnull = df.isnull().sum()
dfnull
```

```python
df.dtypes
```

```python
df['product_type'].unique()
```

```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
df['transaction_time'] = pd.to_datetime(df['transaction_time'],format= '%H:%M:%S')

df.dtypes
```

```python
df['Month']=df['transaction_date'].dt.month
df.tail()
```

```python
df['Total Sales'] = df['transaction_qty'] * df['unit_price']
df.head()
```

**SALES ANALYSIS**

```python
sales_per_day = df.groupby(df['transaction_date'].dt.date)['Total Sales'].sum().reset_index()
print(sales_per_day)
```

```python
sales_per_producttype = df.groupby(df['product_type'])['Total Sales'].sum().reset_index()
sales_per_productcategory= df.groupby(df['product_category'])['Total Sales'].sum().reset_index()

print(sales_per_productcategory)
```

```python
avg_sales_per_order = df['Total Sales'].mean()
avg_sales_per_order
```

```python
df['Hour'] = df['transaction_time'].dt.hour

sales_per_hour = df.groupby(df['Hour'])['Total Sales'].sum().sort_values(ascending=False).reset_index()
sales_per_hour
```

```python
quantity_per_producttype = df.groupby(df['product_type'])['transaction_qty'].sum().reset_index()
quantity_per_productcategory = df.groupby(df['product_category'])['transaction_qty'].sum().sort_values( ascending=False).reset_index()
#print(quantity_per_productcategory )
print(quantity_per_producttype)
```

```python
sales_by_location= df.groupby(df['store_location'])['Total Sales'].sum().sort_values(ascending=False).reset_index()
sales_by_location
```

```python
sales_per_day.to_csv('sales_per_day.csv', index=False)
sales_per_producttype.to_csv('sales_per_producttype.csv', index=False)
sales_per_productcategory.to_csv('sales_per_productcategory.csv', index=False)
sales_per_hour.to_csv ('sales_per_hour.csv',index=False)
quantity_per_producttype.to_csv('quantity_per_producttype.csv', index = False)
quantity_per_productcategory.to_csv('quantity_per_productcategory.csv', index = False)
```
